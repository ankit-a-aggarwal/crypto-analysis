```python
import pandas as pd

df1 = pd.read_csv('ASKS.csv')
df1 = df1.rename(columns={ df1.columns[0]: "id" })
df2 = pd.read_csv('BIDS.csv')
df2 = df2.rename(columns={ df2.columns[0]: "id" })
how1 = ""
if df1.shape[0] > df2.shape[0]:
    how1 = "left"
else:
    how1 = "right"
result = pd.merge(df1, df2, on='id',how=how1)
result = result.drop(['id','coin_y','exchange_y','timestamp_y'], axis=1)
result.to_csv('combine.csv')
print(result)
```
